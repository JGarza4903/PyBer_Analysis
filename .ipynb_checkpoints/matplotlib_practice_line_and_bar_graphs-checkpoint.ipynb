```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add the Legend
plt.legend()
```

```python
#add a new line color, modified line width, and a marker for our data points

# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis, marker='*',color='blue',linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add a grid
plt.grid()

# Add the Legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis, marker='*',color='green', linewidth=2, label = "Boston")

ax.set_xlabel('Date')

ax.set_ylabel('Fare ($)')

ax.grid()
```

```python
#create bargraph plot using ride-sharing data from above
plt.bar(x_axis, y_axis)
```

```python
# create the plot
plt.bar(x_axis,y_axis,color='green',label='Boston')

# create labels for x and y axis
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# create a title
plt.title('PyBerFare by Month')

# Add a legend
plt.legend()
```

```python
# create the plot for horizontal bargraph
plt.barh(x_axis,y_axis)
```

```python
# Create the plot and invert y-axis
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot and invert y-axis
plt.barh(x_axis,y_axis, color='magenta',label = 'Boston')
plt.gca().invert_yaxis()

#adding axis  labels
plt.xlabel('Fare ($)')
plt.ylabel('Date')

#adding Title
plt.title('PyBer Fare by Month')

plt.legend()
```

```python
## Creating a vertical Bar chart with object-oriented interface method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color='cyan',label='Chicago')

ax.set_ylabel('Months')
ax.set_xlabel('Fare ($)')

ax.invert_yaxis()

ax.legend()

ax.set_title('PyBer Fare by Months')
```
