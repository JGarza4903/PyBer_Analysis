```python
import matplotlib.pyplot as plt
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
plt.scatter(x,y, color='magenta', marker = "x", s=x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Averge Fare vs Total Rides')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(x_axis,y_axis)
```

```python

plt.scatter(y_axis,x_axis, color='red',label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
#plt.xlimit(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# iterate through the y-axis data
y_axis_larger=[]

# iterate through the y-axis data
for data in y_axis:
    #multiply each data point in the y-axis by 3 and add it to a new list
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#another awy of doing this replacing the 'for' and 'while' loops
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, linewidth=2, color='xkcd:sky blue', s= y_axis, label='Boston')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')

ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values=[0,0,0,0,0,0,0.2,0,0,0,0,0]
```

```python
#make percentages smaller
plt.subplots(figsize=(8,8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%1.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots(figsize=(8,8))
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
ax.pie(y_axis, labels=x_axis, explode = explode_values, autopct='%.1f%%')
ax = plt.gca().invert_yaxis()
       
plt.show()
```
