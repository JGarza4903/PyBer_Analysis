```python
import matplotlib.pyplot as plt
```

```python
categories = ['pants', 'shirts', 'outerware', 'pajamas', 'underwear', 'shoes']
```

```python
revenue = [25, 15, 5, 20, 10, 25]
```

```python
plt.pie(revenue)
plt.show()
```

```python
plt.pie(revenue, labels=categories)
plt.show()
```

```python
plt.pie(revenue, labels=categories, autopct='%1.1f%%')
plt.show()
```

```python
#Creating list for explode, must be same size as data you are selecting
explode=[0,0.1,0,0,0,0]
```

```python
#Make pie explode
plt.pie(revenue, labels=categories, autopct='%1.1f%%', explode=explode)
plt.show()
```
